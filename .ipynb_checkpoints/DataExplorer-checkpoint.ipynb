## Bureau Balance

```python
bb = pd.read_csv('../data/input_data/bureau_balance.csv')
```

```python
crit_id_bureau = bb["SK_ID_BUREAU"].isin(ids_bureau)
```

```python
test_bb = bb.loc[crit_id_bureau, :]
```

```python
test_bb_agg = test_bb.groupby(["SK_ID_BUREAU", "STATUS"]).size().unstack(fill_value=0)
```

```python
test_bb.to_csv("../data/output_data/test_bb.csv")
test_bb_agg.to_csv("../data/output_data/test_bb_agg.csv")
```

```python
print(bb.shape, test_bb.shape)
```

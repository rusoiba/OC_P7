# Exploratory Data Analysis

## Target Distribution

```python
ax = train["TARGET"].value_counts().plot.bar()
ax.set_title("Target distribution : had payment difficulties [0/1]")
ratio = train["TARGET"].value_counts()[0]/train["TARGET"].value_counts()[1]
print("We have {} times more good loans than bad loans".format(round(ratio, 1)))
```

## Training Variables Distributions

```python
ax = plt.figure(figsize=(18,5))
ax = train.isnull().mean().plot.bar()

print("There are {} variables with no missing values at all".format((train.isnull().mean()==0).sum()))
print("There are {} variables with less than 30% missing values".format((train.isnull().mean()<0.3).sum()))
print("There are {} variables with more than 50% missing values".format((train.isnull().mean()>0.5).sum()))
```
